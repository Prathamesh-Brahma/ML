```python
isa = {
    "bird": "animal",
    "dog": "animal",
    "sparrow": "bird"
}
has = {
    "animal": "cells",
    "dog": "tail"
}
can = {
    "bird": "fly",
    "dog": "bark"
}
def check_isa(x):
    if x in isa:
        print(x, "is a", isa[x])
    else:
        print("No ISA relation found")
def check_has(x):
    if x in has:
        print(x, "has", has[x])
    else:
        print("No HAS relation found")
def check_can(x):
    if x in can:
        print(x, "can", can[x])
    else:
        print("No CAN relation found")
check_isa("sparrow")
check_has("dog")
check_can("bird")
```

```python
#pr9
P_cloudy = 0.40
P_rain = 0.20
P_cloudy_given_rain = 0.85
P_rain_given_cloudy = (P_rain * P_cloudy_given_rain) / P_cloudy
print("Probability of Rain given Cloudy Weather:")
print("P(Rain | Cloudy) =", P_rain_given_cloudy)

print("\nPercentage Probability:")
print("P(Rain | Cloudy) =", P_rain_given_cloudy * 100, "%")
```
